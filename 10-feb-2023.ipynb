```python
# Q3:
# Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
# close the file. Open this file and read the content of the file.

# Open file in write mode
file = open("data_scientist.txt", "w")

# Write a sentence to the file
file.write("I want to become a Data Scientist")

# Close the file
file.close()

# Open the file in read mode
file = open("data_scientist.txt", "r")

# Read the contents of the file
content = file.read()

# Print the contents of the file
print(content)

# Close the file
file.close()
```

```python
# Q4:
# Explain the following with python code: read(), readline() and readlines().
"""
Ans: read() - The read() method reads the entire content of the file as a single string.
"""
# Open a file for reading
file = open("my_file.txt", "r")

# Read the entire content of the file
content = file.read()

# Close the file
file.close()

# Print the content of the file
print(content)



"""
Ans: readline() - The readline() method reads a single line from the file and returns it as a string.
"""
# Open a file for reading
file = open("my_file.txt", "r")

# Read the first line of the file
line = file.readline()

# Close the file
file.close()

# Print the first line of the file
print(line)
readlines() - The readlines() method reads all the lines from the file and returns them as a list of strings.
python
Copy code
# Open a file for reading
file = open("my_file.txt", "r")

# Read all the lines of the file
lines = file.readlines()

# Close the file
file.close()

# Print all the lines of the file
print(lines)


"""
Ans: readlines() - The readlines() method reads all the lines from the file and 
returns them as a list of strings.
"""
# Open a file for reading
file = open("my_file.txt", "r")

# Read all the lines of the file
lines = file.readlines()

# Close the file
file.close()

# Print all the lines of the file
print(lines)
```
