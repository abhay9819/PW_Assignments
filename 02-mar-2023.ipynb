```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title("Scatter plot of x and y")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show the plot
plt.show()

```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
'''
The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows us to divide 
the figure into a grid of subplots and specify which subplot we want to create the plot on.
'''


import numpy as np
import matplotlib.pyplot as plt

# Create data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the subplots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot the lines on the subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to the subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Add x and y labels to the subplots
axs[1, 0].set_xlabel('X values')
axs[1, 1].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[1, 0].set_ylabel('Y values')

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
'''
A box plot, also known as a box-and-whisker plot, is a way to visualize the distribution of a dataset through
their quartiles. It displays a summary of the minimum, maximum, median, first quartile (25th percentile), and 
third quartile (75th percentile) of a dataset. Box plots are useful for detecting outliers and comparing 
multiple datasets.
'''

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xticks([1, 2], ["Box 1", "Box 2"])
plt.ylabel("Value")
plt.show()

```
