```python
# Q5:
# Explain how you can use the find() method to query the MongoDB database. Write a simple code to
# demonstrate this.
"""
The find() method in MongoDB is used to query the database and retrieve documents 
that match a specified query filter. The method returns a cursor object, which can be 
used to iterate over the matching documents.
"""

# Example:
import pymongo

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Select the database and collection
db = client["mydatabase"]
col = db["customers"]

# Define the query filter
myquery = { "address": { "$regex": "^S" } }

# Find the matching documents
mydocs = col.find(myquery)

# Print the matching documents
for x in mydocs:
    print(x)
```

```python
# Q6:
# Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
"""
The sort() method in MongoDB is used to sort the results of a query based on one or more fields.
The method takes one or more sort criteria as arguments, which can be specified as a dictionary
object with field names as keys and sort directions as values.
"""

# Example:
import pymongo

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Select the database and collection
db = client["mydatabase"]
col = db["customers"]

# Sort the documents by name
mydocs = col.find().sort("name", pymongo.ASCENDING)

# Print the sorted documents
for x in mydocs:
    print(x)
```
